```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?
air_bnb.groupby('neighbourhood_group').count()
# 5-available Manhattan-highest frequency
```

```python
# Are private rooms the most popular in manhattan?
air_bnb.groupby(['neighbourhood_group', 'room_type']).count()
# apt are the most popular. there are 13000+ ids of them
```

```python
# Which hosts are the busiest and based on their reviews?
air_bnb.groupby('host_name').sum()
# Cil is the busiest
```

```python
#Which neighorhood group has the highest average price?
air_bnb.groupby('neighbourhood_group').mean().sort_values('price', ascending = True)['price']
# manhattan is the highest price at around 197
```

```python
# Which neighbor hood group has the highest total price?
air_bnb.groupby('neighbourhood_group').sum()
#manhattan has highest total at around 4200000
```

```python
#Which top 5 hosts have the highest total price?
air_bnb.groupby('host_name').sum()

#현선 (Email hidden by Airbnb) 단비  HENRY LEE Queens Corner Lot
```
